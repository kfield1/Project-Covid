```python
import os
import glob
import pandas as pd
#set working directory
#os.chdir("Jan to Feb 2021")

```

```python
#find all csv files in the folder 
#use glob pattern matching -> extension = 'csv'
#save result in list ->all_filenames

os.chdir("Oct to Dec 2020")
extension = 'csv'
all_filesnames20 = ["Oct to Dec 2020/"+i for i in glob.glob('*.{}'.format(extension))]

all_filesnames20
```

```python
os.chdir("../Jan to Feb 2021")
extension = 'csv'
all_filesnames21 = ["Jan to Feb 2021/"+i for i in glob.glob('*.{}'.format(extension))]

os.chdir("..")

all_filesnames21
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filesnames20+all_filesnames21])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')

us_cases_csv=(combined_csv.to_csv)
```
